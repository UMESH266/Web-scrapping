### Webscrapping linkedin for job listings

```python
# Importing necessary libraries
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
```

#### Initialization of webdriver with customized options

```python
# Set up Chrome options to customize the browser behavior
options=webdriver.ChromeOptions()

# Set user-agent to mimic a browser behavior
options.add_argument('user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:126.0) Gecko/20100101 Firefox/126.0')

# Initialization of webdriver 
driver = webdriver.Chrome(options=options)
```

#### Finding and loading of webpage

```python
# URL of the linkedin page
url = "https://www.linkedin.com/jobs/search/?currentJobId=3903646649&geoId=105214831&keywords=junior%20%2F%20associate%20data%20scientist&location=Bengaluru%2C%20Karnataka%2C%20India&origin=JOB_SEARCH_PAGE_SEARCH_BUTTON&refresh=true"

# Open the URL
r = requests.get(url)

# Sleep for 5 seconds to ensure the page loads completely
# sleep(5)

# Get the HTML source code of the page after it has fully loaded
# html = driver.page_source
html = r.content
```

```python
soup = BeautifulSoup(html, 'html5lib')
```
